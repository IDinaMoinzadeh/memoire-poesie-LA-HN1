```python
import requests
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0 (compatible; recherche-academique/1.0)"}

def extraire_poetes(soup, annee):
    resultats = []
    guest_editor = ""
    for tag in soup.find_all(["h2", "h3", "p", "div"]):
        texte = tag.get_text(strip=True)
        if "Guest Editor" in texte:
            guest_editor = texte.replace("Guest Editor:", "").strip()
            break
    texte_complet = soup.get_text("\n")
    lignes = [l.strip() for l in texte_complet.split("\n") if l.strip()]
    i = 0
    while i < len(lignes):
        ligne = lignes[i]
        if i + 1 < len(lignes) and "Appeared in:" in lignes[i + 1]:
            if "," in ligne:
                parties = ligne.split(",", 1)
                nom_poete = parties[0].strip()
                titre_poeme = parties[1].strip()
            else:
                nom_poete = ligne
                titre_poeme = ""
            revue = lignes[i + 1].replace("Appeared in:", "").strip()
            resultats.append({
                "annee": annee,
                "guest_editor": guest_editor,
                "nom_poete": nom_poete,
                "titre_poeme": titre_poeme,
                "revue": revue
            })
            i += 2
        else:
            i += 1
    return resultats

# Test sur 2008 seulement
url = "https://web.archive.org/web/20250420030249/https://bestamericanpoetry.com/archive/?id=22"
r = requests.get(url, headers=headers, timeout=15)
soup = BeautifulSoup(r.text, "html.parser")
poetes = extraire_poetes(soup, "2008")
print(f"{len(poetes)} poètes extraits")
for p in poetes[:5]:
    print(p)
```

```python
def extraire_poetes_v2(soup, annee):
    resultats = []
    
    # Guest editor
    guest_editor = ""
    for tag in soup.find_all(["h2", "h3", "p", "div"]):
        texte = tag.get_text(strip=True)
        if "Guest Editor" in texte:
            guest_editor = texte.replace("Guest Editor:", "").strip()
            break

    texte_complet = soup.get_text("\n")
    lignes = [l.strip() for l in texte_complet.split("\n") if l.strip()]

    for i, ligne in enumerate(lignes):
        if ligne == "Appeared in:" and i >= 2:
            nom_brut = lignes[i - 2]  # "Tom Andrews,"
            titre = lignes[i - 1]     # "Evening Song"
            revue = lignes[i + 1] if i + 1 < len(lignes) else ""
            
            # Nettoyer la virgule finale du nom
            nom_poete = nom_brut.rstrip(",").strip()
            
            resultats.append({
                "annee": annee,
                "guest_editor": guest_editor,
                "nom_poete": nom_poete,
                "titre_poeme": titre,
                "revue": revue
            })

    return resultats

# Test sur 2008
poetes = extraire_poetes_v2(soup, "2008")
print(f"{len(poetes)} poètes extraits")
for p in poetes[:5]:
    print(p)
```

```python
#let's go
import csv
import time

OUTPUT_FILE = "/home/dina/Documents/MEMOIRE BIBLIO/Anthologies nationales/bap_poetes_brut.csv"
DELAI_SECONDES = 5

VOLUMES_MANQUANTS = [
    {"annee": "2008", "id": 22},
    {"annee": "2009", "id": 23},
    {"annee": "2010", "id": 24},
    {"annee": "2011", "id": 25},
    {"annee": "2012", "id": 26},
    {"annee": "2013", "id": 27},
    {"annee": "2014", "id": 28},
    {"annee": "2015", "id": 29},
    {"annee": "2016", "id": 30},
    {"annee": "2017", "id": 31},
    {"annee": "2018", "id": 32},
    {"annee": "2019", "id": 33},
    {"annee": "2020", "id": 34},
    {"annee": "2021", "id": 35},
    {"annee": "2022", "id": 36},
    {"annee": "2023", "id": 37},
    {"annee": "2024", "id": 38},
]

total = 0

with open(OUTPUT_FILE, "a", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["annee", "guest_editor", "nom_poete", "titre_poeme", "revue"])

    for vol in VOLUMES_MANQUANTS:
        url = f"https://web.archive.org/web/20250420030249/https://bestamericanpoetry.com/archive/?id={vol['id']}"
        try:
            r = requests.get(url, headers=headers, timeout=15)
            r.raise_for_status()
            soup = BeautifulSoup(r.text, "html.parser")
            poetes = extraire_poetes_v2(soup, vol["annee"])
            writer.writerows(poetes)
            total += len(poetes)
            print(f"{vol['annee']} : {len(poetes)} poètes")
        except Exception as e:
            print(f"{vol['annee']} : ERREUR — {e}")
        time.sleep(DELAI_SECONDES)

print(f"\nTotal ajouté : {total} entrées")
```
